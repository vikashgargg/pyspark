# Pyspark

### PySpark is an interface for Apache Spark in Python. It not only allows you to write Spark applications using Python APIs, but also provides the PySpark shell for interactively analyzing your data in a distributed environment.  It is majorly used for processing structured and semi-structured datasets. It also provides an optimized API that can read the data from the various data source containing different files formats. Thus, with PySpark you can process the data by making use of SQL as well as HiveQL

```python
import pyspark
```

```python
import pandas as pd
type(pd.read_csv('house price.csv'))
```

```python
from pyspark.sql import SparkSession
```

```python
spark=SparkSession.builder.appName('house_price').getOrCreate()
```

```python
from pyspark.sql.functions import *
```

```python
df=spark.read.csv('house price.csv')
```

```python
### Saleprice of the house less than or equal to 20000
df4=df3.filter("SalePrice>=165500")
```

```python
df4.toPandas().head()
```

```python
#filter with a condition and selecting a particular column
df3.filter("SalePrice<=165500").select(['SaleCondition','YrSold']).show()
```

```python
df3.filter(df3['SalePrice']<=165500).show()
```

```python
df3.filter((df3['SalePrice']<=165500) | 
                  (df3['SalePrice']>=250000)).show()
```

```python
df3.filter(~(df3['SalePrice']<=165500)).show()
```

```python
## Groupby
### Grouped to find the maximum saleprice
df3.groupBy('SaleCondition').sum().show()
```

```python
## Groupby
### Grouped to find the avg saleprice
df3.groupBy('SaleCondition').avg().show()
```

```python
## Groupby
### Grouped to find the mean saleprice
df3.groupBy('SaleCondition').mean().show()
```

```python
## Groupby
### Grouped to find the sum saleprice
df3.groupBy('GarageType').sum().show()
```

```python
## Groupby
### Grouped to find the count saleprice
df3.groupBy('SaleCondition').count().show()
```
