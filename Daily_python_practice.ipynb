```python
d=dict()
for line in data1:
    line=line.strip()
    line=line.lower()
    words=line.split(' ')
    for word in words:
        #print(word)
        if word in d:
            d[word]=d[word]+1
        else:
            d[word]=1

# for key in list(d.keys()):
#     print(key,':',d[key])
```

```python
#Validate password 
import re
password = str(input())
flag = 0
while True:
    if (len(password)<8):
        flag = -1
        break
    elif not re.search("[a-z]", password):
        flag = -1
        break
    elif not re.search("[A-Z]", password):
        flag = -1
        break
    elif not re.search("[0-9]", password):
        flag = -1
        break
    elif not re.search("[_@$]", password):
        flag = -1
        break
    elif re.search("\s", password):
        flag = -1
        break
    else:
        flag = 0
        print("Valid Password")
        break

if flag ==-1:
    print("Not a Valid Password")

```
